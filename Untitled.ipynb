# Data Preparation

```python
# Converting the date column to to date time format
df['date'] = pd.to_datetime(df['date'])
type(df['date'][0])
```

```python
# Setting the Date column as the index
df.set_index('date', inplace= True)
```

```python
# grouping the data by daily amount of energy usage
daily_data = df['Appliances'].resample('D').sum()
```

```python
# checking if the data is stationary
adf_test = adfuller(daily_data)
print(f'ADF Statistic: {adf_test[0]}')
print(f'p-value: {adf_test[1]}')

# in the case where data is not stationary
if adf_test[1] > 0.05:
    daily_data_diff = daily_data.diff().dropna()
    adf_test_diff = adfuller(daily_data_diff)
    print(f'Differenced ADF Statistic: {adf_test_diff[0]}')
    print(f'Differenced p-value: {adf_test_diff[1]}')
else:
    daily_data_diff = daily_data

# Plotting the ACF and PACF results
plot_acf(daily_data_diff, lags=30)
plot_pacf(daily_data_diff, lags=30)
plt.show()
```

```python
# Fitting the SARIMAX model
model = SARIMAX(daily_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 7)).fit()
print(model.summary())
```
