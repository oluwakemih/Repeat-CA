```python
# Converting the date column to to date time format
df['date'] = pd.to_datetime(df['date'])
type(df['date'][0])
```

```python
# Setting the Date column as the index
df.set_index('date', inplace= True)
```

```python
# grouping the data by daily amount of energy usage
daily_data = df['Appliances'].resample('D').sum()
```

```python
# checking if the data is stationary
adf_test = adfuller(daily_data)
print(f'ADF Statistic: {adf_test[0]}')
print(f'p-value: {adf_test[1]}')

# in the case where data is not stationary
if adf_test[1] > 0.05:
    daily_data_diff = daily_data.diff().dropna()
    adf_test_diff = adfuller(daily_data_diff)
    print(f'Differenced ADF Statistic: {adf_test_diff[0]}')
    print(f'Differenced p-value: {adf_test_diff[1]}')
else:
    daily_data_diff = daily_data

# Plotting the ACF and PACF results
plot_acf(daily_data_diff, lags=30)
plot_pacf(daily_data_diff, lags=30)
plt.show()
```

```python
# Fitting the SARIMAX model
model = SARIMAX(daily_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 7)).fit()
print(model.summary())
```

```python
# forcast
forecast_steps = 30
forecast_values = model.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=daily_data.index[-1] + pd.DateOffset(days=1), periods=forecast_steps, freq='D')
forecast_series = pd.Series(forecast_values.predicted_mean, index=forecast_index)

# Plotting the forecast
plt.figure(figsize = (12, 6))
plt.plot(daily_data, label = 'Observed')
plt.plot(forecast_series, label = 'Forecast')
plt.legend()
plt.show()
```

```python
# Aligning the actual and forecast data for error calculation
actual = daily_data[-forecast_steps:].copy()
forecast_series = forecast_series[:len(actual)]
actual.index = forecast_series.index

# Calculating forecast errors
mae = mean_absolute_error(actual, forecast_series)
rmse = np.sqrt(mean_squared_error(actual, forecast_series))
print(f'Test MAE: {mae}, Test RMSE: {rmse}')
```

```python
# Dataset Rows and Column counts
print('Number of Columns:', data.shape[1])
print('Number of Rows:', data.shape[0])
print('Null Values:', data.isna().sum().sum())
print('Duplicates:', data.duplicated(keep = 'first').sum())
```

```python
basket = (data
          .groupby(['Member_number', 'itemDescription'])['itemDescription']
          .count().unstack().reset_index().fillna(0)
          .set_index('Member_number'))

# Convert quantities to a binary format
def encode_units(x):
    return 1 if x >= 1 else 0

basket_sets = basket.applymap(encode_units)
```

```python
# Apriori algorithm
start = time.time()
frequent_itemsets_apriori = apriori(basket_sets, min_support=0.01, use_colnames=True)
end = time.time()
rules_apriori = association_rules(frequent_itemsets_apriori, metric='lift', min_threshold=1)
print(f'Time taken by Apriori:{end-start} seconds')
```

```python
# Apriori algorithm
start = time.time()
frequent_itemsets_fpgrowth = fpgrowth(basket_sets, min_support=0.01, use_colnames=True)
end = time.time()
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric='lift', min_threshold=1)
print(f'Time taken by FP-Growth:{end-start} seconds')
```

```python
# Identifying the top 10 most Frequently purchased items
top_10_items = basket_sets.sum().sort_values( ascending=False).head(5)
print('Top 10 Most Frequently Purchased Items:')
print(top_10_items)
```

```python
# Extracting the top 5 association rules based on lift
top_5_rules_apriori = rules_apriori.sort_values(by='lift', ascending=False).head(5)
print('\nTop 5 Association Rules (Apriori):')
print(top_5_rules_apriori)
```

```python
# Extracting the top 5 association rules based on lift
top_5_rules_fpgrowth = rules_fpgrowth.sort_values(by='lift', ascending=False).head(5)
print('\nTop 5 Association Rules (FP-Gowth):')
print(top_5_rules_fpgrowth)
```
